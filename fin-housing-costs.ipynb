```python
#***************NEW COLUMN IS ADDED AND RENAMED************************.
df.rename(columns={'area':'size', 'region':'localArea', 'maakunta':'region'}, inplace=True)
df['price/m2'] = df['price']/df['size']
df.head(4)
```

*****************CHANGE YEAR FROM STR TO INT*********************************************

```python
##there are "-" year which will not convert to int by simple conversion test_df['year'] = test_df['year'].astype('Int64') 
df['year'] = pd.to_numeric(df['year'], errors='coerce') ## it will convert to int/float and NaN if there are non-numeric str(here "-")
df['year'] = df['year'].astype('Int64') ## still make sure that all nubers are int
```

****************REPLACING LARGER OUTLIERS WITH NaN***********************
Ther were some serious ouliers in the "price/m2" due to division. Dividing with Nan, zeros and some outliers produced,
infinte values and outliers from 20000 upto 500000 euros per m2, which is impossible. 

```python
import numpy as np
#************************REPLACING LARGER OUTLIERS WITH NaN***********************
df.replace([np.inf, -np.inf], np.nan, inplace=True) #replacing infinite value
df['price/m2'] = df['price/m2'].mask(df['price/m2']>15000, None) # replacing large outliers OR df.loc[df['col1'] > 10000, 'col1'] = None 
sns.boxplot(data=df['price/m2'])
df['price/m2'].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
filt_size = df['size']<300
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,3))
sns.histplot(data=df, x='price', kde=True, ax=axes[0])
sns.histplot(data=df[filt_size], x='size', kde=True, ax=axes[1])
sns.histplot(data=df, x='price/m2', kde=True, ax=axes[2])

plt.tight_layout()
```

```python
#gr = df.groupby(df['region'])
#gr_df=gr['price'].median().sort_values(ascending=False).reset_index()
no_ah_df = df[df['region']!='Ahvenanmaa'] # df without Ahvenmaa because ahvenmaa have only 5 data and misleading graphs

fig, axe = plt.subplots(3,1,figsize=(10, 6),sharex=True)
sns.barplot(data = no_ah_df, y='price', x='region', estimator='median', ax=axe[0])
axe[0].set_ylabel('median price')


# Create a second y-axis (ax2) that shares the same x-axis
ax2 = axe[0].twinx()
# Plot a lineplot on the second y-axis (ax2)
sns.lineplot(data = no_ah_df, y='price/m2', x='region', estimator='median', ax=ax2, color='red', label='median price/m2')
# Set labels for the second y-axis
ax2.set_ylabel('median price/m2')
ax2.tick_params(axis='y', labelcolor='red')


percent_df = df['region'].value_counts(normalize=True).reset_index()
sns.barplot(data = no_ah_df, y='size', x='region', estimator='median', ax=axe[1] )
axe[1].set_ylabel('median size')

sns.barplot(data = percent_df, y='proportion', x='region', ax=axe[2] )

axe[2].set_xticklabels(axe[2].get_xticklabels(), rotation=90)
#plt.tight_layout()
plt.show()
```
